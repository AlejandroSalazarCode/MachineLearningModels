```python
# 0 = Suspension, 1 = Electrical, -1 = Abstain/Unknown
@labeling_function()
def keyword_suspension(x):
    return 0 if "shock absorber" in x.text.lower() else -1

@labeling_function()
def keyword_electrical(x):
    return 1 if "fuse box" in x.text.lower() or "alternator" in x.text.lower() else -1
```

```python
lfs = [keyword_suspension, keyword_electrical]
applier = PandasLFApplier(lfs=lfs)

# Define df_train with some sample data to resolve the NameError
data = {'text': ['This car has a broken shock absorber.', 'I need to check the fuse box.', 'The alternator is not working.']}
df_train = pd.DataFrame(data)

# L_train is a matrix of the generated labels which you can then use
# to train a standard Scikit-Learn text classifier!
L_train = applier.apply(df_train)
```
